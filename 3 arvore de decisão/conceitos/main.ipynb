```python
from sklearn.tree import DecisionTreeClassifier
```

Base risco de crédito

```python
import pickle
with open(r'D:\Biblioteca\Documents\Machine_Learning\1 pre_processamento_dados\base_final\risco_credito.pkl', 'rb') as f:
    X_risco_credito, y_risco_credito = pickle.load(f)
```

```python
arvore_risco_credito = DecisionTreeClassifier(criterion='entropy')
arvore_risco_credito.fit(X_risco_credito, y_risco_credito)
```

Aqui seria o nivel de importancia que é usada na construção da arvore de decisao

```python
arvore_risco_credito.feature_importances_
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
# nome das colunas
previsores = ['história', 'dívida', 'garantias', 'renda']
# ajustando o tamanho do grafico
figura, eixo = plt.subplots(nrows=1, ncols=1, figsize=(10,10))
tree.plot_tree(arvore_risco_credito, feature_names=previsores, class_names=arvore_risco_credito.classes_, filled=True);
```
