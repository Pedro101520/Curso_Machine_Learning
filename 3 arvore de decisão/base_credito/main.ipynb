```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
import pickle
with open(r'D:\Biblioteca\Documents\Machine_Learning\1 pre_processamento_dados\base_final\credit.pkl', 'rb') as f:
    X_credit_treinamento, y_credit_treinamento, X_credit_teste, y_credit_teste = pickle.load(f)
```

```python
arvore_credit = DecisionTreeClassifier(criterion='entropy')
arvore_credit.fit(X_credit_treinamento, y_credit_treinamento)
```

```python
previsoes = arvore_credit.predict(X_credit_teste)
previsoes
```

```python
print("Acurácia:", accuracy_score(y_credit_teste, previsoes))
```
